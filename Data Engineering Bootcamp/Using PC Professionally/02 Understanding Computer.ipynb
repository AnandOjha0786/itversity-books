### Components of Computer
Components of Computer can be broadly categorized into two – Hardware and Software.

**Hardware Components**

Let us start with Hardware Components.

* **Memory** –  Good to have 8GB or 16GB RAM.
* **CPU** – Processor can be dual core or quad core.Quad core has double the speed than dual core.16GB with quad core is an ideal combination.
* **Storage** – Can be SSD or Hard disk.Storage doesn’t affect the performance.Data can be stored on cloud as well.
* **Networking** – Better internet connectivity will ensure good performance.

## Profile and Environment variables
Once you login to the system, you will land into your desktop. Desktop is personalized entry point into a computer. You can change it as per your requirements using profile and environment variables.

***Setting environment variables in Linux***

* Open either Mac or Linux terminal.This will take you to home directory.
* Use the commands <mark>ls -altr</mark> to list out all the files, here you will see **.profile** and .bash_profile.  .profile is used by the user to have some specific settings for his own account.
* The .profile and .bashrc files controls all the environment settings.
* To set PATH in mac, open .profile and append to it – <mark>export PATH=/usr/local/jdk1.7.0/bin:$PATH</mark>
* Never replace the contents of the path, always append to it.

***Setting environment variables in Windows***

* Go to Environment Variables and then choose System Variables unless explicitly specified as part of instructions.
* There are two types of environment variables  –  **_HOME** and **PATH**
* **_HOME** is set to base directory of the application. Base directory is nothing but main directory into which files related to the software installed is copied.
* The **PATH** is the system variable that your operating system uses to locate needed executables from the command line or Terminal window.
* Executables will be typically under bin directory.

***Example***

Let us download and install Java and understand what it actually mean. We will see both on Windows or Linux (if you are using Mac you can follow Linux instructions)

* Make sure if Java is already installed on your PC – <mark>java -version</mark>
* Search for JDK and download fro
* Go to file explorer on Windows – C:\Program Files\Java\jdk
* On the Advanced tab, select Environment Variables, and then edit JAVA_HOME to point to this location.
* So, typically we will be setting JAVA_HOME to base directory which also has many sub-directories including bin directory.
* We need to update PATH variable to bin directory to get executables to run.
* Set the <mark>PATH</mark> environment variable if you want to be able to easily run the executables (<mark>javac.exe, java.exe, javadoc.exe etc.</mark>) from any directory without having to type the full path of the command. In case, the <mark>PATH</mark> variable is not set, you need to specify the full path to the executable every time you run it on terminal – C:\Java\jdk1.7.0\bin\javac MyClass.java
* Now, try running javac from terminal after setting the PATH.

In some cases installation will take care of updating global profiles and registries (in windows).We need not configure anything.
* You can validate following these instructions
 * <mark>echo $PATH</mark>– This will give you the directory path that is set in PATH variable. For Windows, you can say <mark>echo %PATH%</mark>
 * <mark>which -ls</mark> – It returns the pathnames of the files which would be executed in the current environment (on Mac or Linux)

## Basic and Advanced editors

Quite often we might have to edit some plain text files. Here are some of the tips related to editing text files.

* People tend to use notepad or Microsoft Word for editing text files.
 * Notepad is very primitive and it is counter productive.
 * Microsoft Word is not for plain text, it is primarily to create files with images (images are not plain text)
* Do not use notepad or Microsoft Word for text editing.
* If it is code better to use IDEs (advanced editors and specific to programming languages)
* If it is just plain text use standard text editors. Here are some of the features of standard text editors.
 * Support for regular expressions
 * Ability to edit files on remote servers
 * Columnar Selection
 * Macros – for replayable text editing
 * Popular editors are – Notepad++, Ultraedit, Sublime Text
 
We would highly recommend to choose Sublime Text or any one editor of your choice and be expert with it. It might take around 40 hours, but it will save thousands of hours in your career.
* Depending on target servers we need to configure editors to save files in particular modes like DOS mode or unix mode.

## Connecting to remote servers

Now let us understand details related to connecting to remote servers.

* Once the application is deployed in production, it will be supported by L1 support team.
* Developers are considered as L3 support, if L1 support think that it is issue with respect to application then developers need to troubleshoot the issue.
 * Troubleshoot the application failure by analysing logs
 * Data related issues by running queries in the database or by reviewing the logic in the code.
* Developers might have to connect to the production servers with restricted permissions to understand the problem and then reproduce it. Here are different alternatives using which developers can connect to remote servers.
 * ssh tools
 * putty or Cygwin on Windows 8 or earlier. Putty is primitive, better to use terminal or some advanced tools in Windows other than putty.
 * Linux subsystem in Windows 10.
 * Terminal on Mac/Linux
* Keep in mind that once the source for the problem is identified, the issue should be reproduced locally and go through the entire life cycle to fix the issue.
